```python
Auto = pd.read_csv('Auto.csv', na_values=['?'])
Auto = Auto.dropna()
Auto
```

```python
qualitative_columns = ['cylinders', 'origin', 'name']
quantitative_columns = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration', 'year']
```

```python
import numpy as np
for col in quantitative_columns:
    min_val = np.min(Auto[col])
    max_val = np.max(Auto[col])
    print(f"{col}: min = {min_val}, max = {max_val}, range = ({min_val}, {max_val})")
```

```python
for col in quantitative_columns:
    mean = np.mean(Auto[col])
    std = np.std(Auto[col])
    print(f"{col}: mean = {mean}, std = {std}")
```

```python
Auto_removed = Auto.copy().drop(Auto.index[9:85])
Auto_removed[:20]
```

```python
for col in quantitative_columns:
    min_val = np.min(Auto_removed[col])
    max_val = np.max(Auto_removed[col])
    mean = np.mean(Auto_removed[col])
    std = np.std(Auto_removed[col])
    print(f"{col}: min = {min_val}, max = {max_val}, range = ({min_val}, {max_val}), mean = {mean}, std = {std}")
```
